```python
student = data.drop(['ZIP code (state)'], axis=1)
student = student.drop(['ZIP code (zip)'], axis=1)
student = student.drop(['ZIP code (city)'], axis=1)
student = student.drop(['ZIP code (address)'], axis=1)
```

```python
print(student.head())
```

```python
corelation = student.corr()
```

```python
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot=True)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x= 'Age', y= 'Year', hue= 'Problem Area in Diabetes (PAID) Scale Score', data= student)
```

```python
sns.relplot(x= 'Age', y= 'Year', hue= 'Gender', data= student)
```

```python
sns.relplot(x= 'Tobacco Use (Yes/No)', y= 'Age', hue= 'Diabetes Status (Yes/No)', data= student)
```

```python
sns.relplot(x= 'Sugar-Sweetened Beverage Consumption', y= 'Carbohydrate Counting', hue= 'Diabetes Status (Yes/No)', data= student)
```
